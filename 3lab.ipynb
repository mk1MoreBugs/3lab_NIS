```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
rng = np.random.default_rng()
```

```python
def get_random_array():
    return rng.standard_normal(100) - 2
```

```python
X1 = get_random_array()
```

```python
print(X1)
```

```python
ts1 = X1 * 50 + 1600
```

```python
b = 0.5
X2 = get_random_array()
failure_rate = (2 / 4000) ** b
tv1 = (-1 / failure_rate * np.log(1 - X2)) ** (1 / b)
```

```python
tv1
```

```python
X3 = get_random_array()
failure_rate = 1 / 6000
tp1 = -1 / failure_rate * np.log(1 - X3)
```

```python
tp1
```

```python
T1 = np.array([ts1,tv1,tp1])
```

```python
print(T1.shape)
```

```python
t1 = np.minimum(T1[0], T1[1], T1[2])
```

```python
X4 = get_random_array()
X5 = get_random_array()
X6 = get_random_array()
```

```python
b = 1.5

ts2 = X4 * 100 + 2300 

failure_rate = (0.9033 / 2000) ** b
tv2 = ((-1 / failure_rate * np.log(1 - X5)) ** 2) ** (1/3)

```

```python
tv2
```

```python
failure_rate = 1 / 6000
tp2 = (-1 / failure_rate) * np.log(1 - X6)
```

```python
tp2
```

```python
T2 = np.array([ts2,tv2,tp2])
```

```python
t2 = np.minimum(T2[0], T2[1], T2[2])
```

```python
T12 = np.array([t1, t2])
```

```python
T12_transpose = T12.transpose()

list_max_element = list()
list_max_element_index = list()

for t1_t2_element in T12_transpose:
    max_element = t1_t2_element[0]
    max_element_index = 0
    
    for index in range(1, len(t1_t2_element)):
        if t1_t2_element[index] > max_element:
            max_element = t1_t2_element[index]
            max_element_index = index
            
    list_max_element.append(max_element)
    list_max_element_index.append(max_element_index + 1)
```

```python
plt.plot(list_max_element_index, 'ro')
plt.title('o')
plt.show()
```

```python
plt.plot(list_max_element)
plt.show()
```

```python
tm = np.mean(list_max_element)
```

```python
t = np.sort(list_max_element)
```

```python
N0 = 100
Nt = [i for i in range(N0, 0, -1)]
```

```python
print(Nt)
```

```python
def calculate_probability(element):
    return element / N0

p = list(map(calculate_probability, Nt))
```

```python
plt.plot(t, p)
plt.xlabel('t')
plt.ylabel('p')
plt.show()
```
